### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



Verificación de la Estructura de Datos

```python
print("--- Tienda 1 (tienda) Info ---")
tienda.info()
print("\n--- Tienda 2 (tienda2) Info ---")
tienda2.info()
print("\n--- Tienda 3 (tienda3) Info ---")
tienda3.info()
print("\n--- Tienda 4 (tienda4) Info ---")
tienda4.info()
```

Calculo de la facturación de cada tienda , suma total de las ventas

```python
# Calcular la facturación total (suma de 'Precio') para cada tienda

facturacion_t1 = tienda['Precio'].sum()
facturacion_t2 = tienda2['Precio'].sum()
facturacion_t3 = tienda3['Precio'].sum()
facturacion_t4 = tienda4['Precio'].sum()

# Mostrar los resultados

print(f"\nFacturación Total Tienda 1: ${facturacion_t1:,.2f}")
print(f"Facturación Total Tienda 2: ${facturacion_t2:,.2f}")
print(f"Facturación Total Tienda 3: ${facturacion_t3:,.2f}")
print(f"Facturación Total Tienda 4: ${facturacion_t4:,.2f}")
```

# 2. Ventas por categoría

```python
# Categorías Más Vendidas (Tienda 1)

print("\n--- Categorías Más Vendidas (Tienda 1) ---")
todas_categorias_t1 = tienda['Categoría del Producto'].value_counts()
print(todas_categorias_t1)

# Categorías Más Vendidas (Tienda 2)

print("\n--- Categorías Más Vendidas (Tienda 2) ---")
todas_categorias_t2 = tienda2['Categoría del Producto'].value_counts()
print(todas_categorias_t2)

#  Categorías Más Vendidas (Tienda 3)

print("\n--- Categorías Más Vendidas (Tienda 3) ---")
todas_categorias_t3 = tienda3['Categoría del Producto'].value_counts()
print(todas_categorias_t3)

#  Categorías Más Vendidas (Tienda 4)

print("\n--- Categorías Más Vendidas (Tienda 4) ---")
todas_categorias_t4 = tienda4['Categoría del Producto'].value_counts()
print(todas_categorias_t4)
```

# 3. Calificación promedio de la tienda


Calculamos usando .mean() en la columna Calificador.

```python
# Calcular la calificación promedio para cada tienda

calificacion_t1 = tienda['Calificación'].mean()
calificacion_t2 = tienda2['Calificación'].mean()
calificacion_t3 = tienda3['Calificación'].mean()
calificacion_t4 = tienda4['Calificación'].mean()

# resultados

print(f"\nCalificación Promedio Tienda 1: {calificacion_t1:.2f}")
print(f"Calificación Promedio Tienda 2: {calificacion_t2:.2f}")
print(f"Calificación Promedio Tienda 3: {calificacion_t3:.2f}")
print(f"Calificación Promedio Tienda 4: {calificacion_t4:.2f}")
```

# 4. Productos más y menos vendidos

Se utilizará el metodo .value_counts().

El producto más vendido estará al inicio y el producto menos vendido estará al final.

```python
# Lista Completa de Productos Más Vendidos (Tienda 1)

print("\n Lista Completa de Productos Más Vendidos (Tienda 1)")
productos_t1 = tienda['Producto'].value_counts()
print(productos_t1)

# Lista Completa de Productos Más Vendidos (Tienda 2)

print("\n Lista Completa de Productos Más Vendidos (Tienda 2)")
productos_t2 = tienda2['Producto'].value_counts()
print(productos_t2)

# Lista Completa de Productos Más Vendidos (Tienda 3)

print("\n Lista Completa de Productos Más Vendidos (Tienda 3)")
productos_t3 = tienda3['Producto'].value_counts()
print(productos_t3)

# Lista Completa de Productos Más Vendidos (Tienda 4)

print("\n Lista Completa de Productos Más Vendidos (Tienda 4)")
productos_t4 = tienda4['Producto'].value_counts()
print(productos_t4)
```

# 5. Envío promedio por tienda

```python
# Calcular el costo promedio de envío para cada tienda usando mean()

envio_promedio_t1 = tienda['Costo de envío'].mean()
envio_promedio_t2 = tienda2['Costo de envío'].mean()
envio_promedio_t3 = tienda3['Costo de envío'].mean()
envio_promedio_t4 = tienda4['Costo de envío'].mean()

# Resultados

print(f"\nCosto Promedio de Envío Tienda 1: ${envio_promedio_t1:,.2f}")
print(f"Costo Promedio de Envío Tienda 2: ${envio_promedio_t2:,.2f}")
print(f"Costo Promedio de Envío Tienda 3: ${envio_promedio_t3:,.2f}")
print(f"Costo Promedio de Envío Tienda 4: ${envio_promedio_t4:,.2f}")
```

# 6. Metricas de rendimiento




Cálculos


```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear un diccionario con los resultados

facturacion_t4_ej = 1038376000.0
facturacion_t3_ej = 1098020000.0
facturacion_t2_ej = 1116344000.0
facturacion_t1_ej = 1150880000.0

datos_rendimiento = {
    'Tienda': ['Tienda 4', 'Tienda 3', 'Tienda 2', 'Tienda 1'],

    # La facturacion_tX es el resultado de la suma

    'Facturacion_Total': [facturacion_t4_ej, facturacion_t3_ej, facturacion_t2_ej, facturacion_t1_ej],
    'Calificacion_Promedio': [3.995759, 4.048326, 4.037304, 3.976685],
    'Costo_Envio_Promedio': [23459.457167, 24805.680373, 25216.235693, 26018.609580]
}

# Crear el DataFrame de rendimiento

df_rendimiento = pd.DataFrame(datos_rendimiento)
df_rendimiento = df_rendimiento.set_index('Tienda')

# Ordenar la tabla de menor a mayor facturación

df_rendimiento = df_rendimiento.sort_values(by='Facturacion_Total', ascending=True)

pd.options.display.float_format = '{:,.2f}'.format

# 4. Redondear las columnas de promedio

df_rendimiento['Calificacion_Promedio'] = df_rendimiento['Calificacion_Promedio'].round(2)
df_rendimiento['Costo_Envio_Promedio'] = df_rendimiento['Costo_Envio_Promedio'].round(2)


print("\n--- Tabla de Rendimiento Consolidado (Formato Legible) ---")
print(df_rendimiento)
```

Gráficos


1. Facturación total

```python
# Gráfico 1: Facturación Total

plt.figure(figsize=(10, 6))
bars = plt.bar(
    df_rendimiento.index,
    df_rendimiento['Facturacion_Total'],
    color=['red', 'orange', 'gold', 'green'],
    edgecolor='black'
)

plt.title('Gráfico 1: Facturación Total por Tienda, de menor a mayor')
plt.ylabel('Facturación Total')
plt.xlabel('Tienda')
plt.xticks(rotation=0)
plt.ticklabel_format(style='plain', axis='y')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Valores encima de cada barra

for bar in bars:
    yval = bar.get_height()

    plt.text(bar.get_x() + bar.get_width()/2, yval + 1e7, f'${yval:,.0f}',
             ha='center', va='bottom', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```

2. Calificación promedio

```python
# Gráfico 2: Calificación Promedio

df_calificacion = df_rendimiento.sort_values(by='Calificacion_Promedio', ascending=True)

# Colores

colores_para_calificacion = ['red', 'orange', 'gold', 'green']


# Gráfico 2 eje Y

plt.figure(figsize=(10, 6))
bars2 = plt.bar(
    df_calificacion.index,
    df_calificacion['Calificacion_Promedio'],
    color=colores_para_calificacion,
    edgecolor='black'
)

plt.title('Gráfico 2: Calificación Promedio')
plt.ylabel('Calificación Promedio (1 a 5)')
plt.xlabel('Tienda')
plt.ylim(3.9, 4.1)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Añade los valores encima de cada barra

for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, f'{yval:.2f}',
             ha='center', va='bottom', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```

3. Costo de envío promedio

```python
# Gráfico 3

plt.figure(figsize=(10, 6))
bars3 = plt.bar(
    df_rendimiento.index,
    df_rendimiento['Costo_Envio_Promedio'],
    color=['red', 'orange', 'gold', 'green'],
    edgecolor='black'
)

plt.title('Gráfico 3: Costo de Envío Promedio por Tienda ')
plt.ylabel('Costo de Envío Promedio')
plt.xlabel('Tienda')
plt.xticks(rotation=0)
plt.ticklabel_format(style='plain', axis='y')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Añade los valores encima de cada barra

for bar in bars3:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 500, f'${yval:,.2f}',
             ha='center', va='bottom', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```
